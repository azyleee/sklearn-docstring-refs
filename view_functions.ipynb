```python
import pickle
import pandas as pd

def read_to_df(pkl_path):
    with open(pkl_path, 'rb') as file:
        d = pickle.load(file)
    df = pd.DataFrame.from_dict(d, orient='index')
    filename = pkl_path.replace('.pkl','.xlsx')
    df.to_excel(filename)
    return df
```

```python
read_to_df('functions_info.pkl')
```
